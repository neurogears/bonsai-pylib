```python
import clr
from bonsai_pylib import load_bonsai_config
load_bonsai_config()
import System
from System.Linq.Expressions import Expression

# Add reference to the Harp package
clr.AddReference("Harp.Behavior")
import Harp.Behavior as behavior
clr.AddReference("Bonsai.Harp")
import Bonsai.Harp as harp

from System import Enum

from harp_loader import read_harp_bin
```

```python
behavior.Device.RegisterMap[32].ToString()

a = behavior.Device.RegisterMap[32]
method = (a.GetMethod("GetPayload"))
ret_type = method.ReturnType

def create_parse_reg(register):
    method = (register.GetMethod("GetPayload"))
    ret_type = method.ReturnType
    return lambda x: Enum.ToObject(ret_type, x)
```

```python
file_path = r"data/Streams_34"
data = read_harp_bin(file_path)
data
```

```python
new_df = data.applymap(create_parse_reg(behavior.Device.RegisterMap[34]))
new_df
```

```python
names = System.Enum.GetNames(behavior.DigitalOutputs)
cols = [name for name in names]
named_col_df = data.copy()
for col in cols:
    parser = lambda x: behavior.DigitalOutputs(x).HasFlag(eval(f"behavior.DigitalOutputs.{col}"))
    named_col_df[col] = data.applymap(parser)
```
